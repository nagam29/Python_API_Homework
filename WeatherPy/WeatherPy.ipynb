```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from scipy import stats

# Import API key
from api_keys import weather_api_key 

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "/Users/mikan/Documents/Data Science Bootcamp/Python_API_Homework/WeatherPy/"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

```python
# Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it'sbeing processed (with the city number and city name).

from pprint import pprint

 # Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
query_url

# How the raw data look like? 
city1='london'
resp1=requests.get(query_url + city1).json()
pprint(resp1)
```

```python
# set up lists to hold reponse info
City_id=[]
City=[]
Cloudiness=[]
Country=[]
Date=[]
Humidity=[]
Lat=[]
Lng=[]
Max_temp=[]
Wind_Speed=[]

# Loop through the list of cities and perform a request for data on each
# Include a print log of each city as it's being processed (with the city number and city name).
for city in cities:
    response = requests.get(query_url + city).json()
    try:
        City_id.append(response['id'])
        City.append(response['name'])
        Cloudiness.append(response['clouds']['all'])
        Country.append(response['sys']['country'])
        Date.append(response['dt'])
        Humidity.append(response['main']['humidity'])
        Lat.append(response['coord']['lat'])
        Lng.append(response['coord']['lon'])
        Max_temp.append(response['main']['temp_max'])
        Wind_Speed.append(response['wind']['speed'])
        print ('City_ID: ' + str(response["id"]), 'City_name: ' + response["name"])
    except KeyError:
        print("The key has an issue.")
        
```

**Goal of the analysis:

The purpose of the analysis below is to find whether the variability of temperature, humidity, cloudiness and wind speed are related to latitudes of the cities. 

**Summary findings: 

The variability in the temperatures of the cities seemed to be related to their latitudes. Other weather factors, humidity, cloudiness and wind speed, might be determined by the locations (latitudes) of the cities to some extent, but the relationships between these 3 factors and the latitudes appeared to be really weak. This indicates that other factors (other than the latitudes) that were not taken into consideration during the analysis below might determine these 3 weather conditions.

In northern hemisphere, the higher the latitudes of the cities are, the lower the temperatures become. The scatter plot, City Latitude vs Max Temperature (F), showed this negative relationship for the cities that are located in the latitudes higher than 0 (the cities in northern hemisphere). The regression analysis on the relationships between temperatures and the locations of the cities in northern hemisphere also presented the similar result. The regression analysis also indicated that the locations of the cities (latitudes) might be one of the strong determinants of the temperature. 

In southern hemisphere, the higher the latitudes of the cites are, the higher the temperatures become.  This positive relationship was observed in the cities that were located in the latitudes lower than 0 (please see City Latitude vs Max Temperature (F) scatter plot below).  This trend was also observed with the regression analysis. Unlike in the cities in northern hemisphere, the latitudes did not appear to be as strong determinant of the temperatures in southern hemisphere, based on the regression analysis.

**Limitations of this analysis:

The difference in the strength of the temperature-latitudes-relationships between northern and southern hemispheres noted above could potentially be created as a function of the numbers of available data points: In general, northern hemisphere has a larger number of cities than southern hemisphere. Thus, a larger number of data points in northern hemisphere were naturally available than the one in southern hemisphere. A sufficiently large sample size in north hemisphere might have resulted in a closer to a normal distribution of the means, which also contributed that the regression line for northern hemisphere fitted the data better than the one for southern hemisphere. 

To address the difference in the availability of data points between northern and southern hemispheres, we might need to sample equal number of cities from each hemisphere.

```python
# Create Northern and Southern Hemisphere DataFrames

# Select rows in Northern Hemisphere
north_df=city_df.loc[city_df['Lat'] >= 0]
north_df.head()

# Select rows in Southern Hemisphere
south_df=city_df.loc[city_df['Lat'] < 0]
south_df.head()
```

```python
# Northern Hemisphere - Temperature (F) vs. Latitude
x_values = north_df['Lat']
y_values = north_df['Max_temp_F']

# Perform a linear regression on temperature vs. latitude
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

# Get regression values
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
```

```python
# Put the steps above into a function

def reg_line(x_values,y_values, z, t):
    # Perform a linear regression on temperature vs. latitude
    (slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

    # Get regression values
    regress_values = x_values * slope + intercept

    line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    print(line_eq)
    
    fig = plt.figure()
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")

    # Label plot and annotate the line equation
    plt.xlabel('Latitude')
    plt.ylabel(z)
    plt.title(t)

    # Print r square value
    print(f"The r-squared is: {rvalue}")

    # Show plot
    plt.show()
    fig.savefig(f"{t}.png")
```

```python
# Northern Hemisphere - Temperature (F) vs. Latitude
reg_line(north_df['Lat'], north_df['Max_temp_F'], 'Temperature (F)','Northern_Hemisphere_Temperature_(F)_vs_Latitude')
```

```python
# Southern Hemisphere - Temperature (F) vs. Latitude
reg_line(south_df['Lat'], south_df['Max_temp_F'], 'Temperature (F)','Southern_Hemisphere_Temperature_(F)_vs_Latitude')
```

Northern and Southern Hemisphere - Temperature (F) vs. Latitude

These regression lines are modeling the relationships between the latitudes of the cities and their max temperatures in northern hemisphere and southern hemisphere. 

In general, the cities in northern hemisphere present the negative relationships between the latitudes and the max temperatures. The temperatures decrease by 1.22F as latitude increases one unit. Based on the r-square (-0.88), the regression model appears to explain about 88% of variability in the temperatures in northern hemisphere. In other words, 88% of variability in the cities’ temperatures are accounted for by the latitudes. This result indicates a strong relationship between  the temerpature and the location of the cities. This results is also supported by that the most of data points are close to the regression line.

The cities in southern hemisphere present positive relationships between their latitudes and max temperatures. The temperatures increase by 0.39F as latitude increases one unit. The r-square (0.51) indicates that the regression model can explain about 51% of variability in the temperatures in southern hemisphere. In southern hemisphere, the relationship between temperature and the locations of the cities may not be as strong as the one in northern hemispher.
