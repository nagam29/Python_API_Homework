```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os


# Import API key
from api_keys import g_key
```

```python
# Store Part I results into DataFrame
# Load the csv exported in Part I to a DataFrame

city_df = pd.read_csv("/Users/mikan/Documents/Data Science Bootcamp/Python_API_Homework/WeatherPy/export_city_df.csv") 
city_df.head()
```

```python
gmaps.configure(api_key=g_key)
```

```python
# Humidity Heatmap
# Configure gmaps.
# Use the Lat and Lng as locations and Humidity as the weight. 
# Add Heatmap layer to map. 

#city_df.dtypes

# Store latitude and longitude in locations
locations = city_df[["Lat", "Lng"]]
locations.head()

# Humidity as the weight
weight_humi=city_df['Humidity']

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=weight_humi, 
                               dissipating=False, max_intensity=300,
                                point_radius=5)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Create new DataFrame fitting weather criteria
# Narrow down the cities to fit weather conditions.
# A max temperature lower than 80 degrees but higher than 70.
# Wind speed less than 10 mph.
# Zero cloudiness.
# Drop any rows that don't contain all three conditions. You want to be sure the weather is ideal.

# Create a variable for temeratures in Fahrenheit 
# reference https://www.pythonforbeginners.com/code-snippets-source-code/python-code-celsius-and-fahrenheit-converter
Max_temp_F=(city_df['Max_temp']* 9/5) + 32

# Add this field to city_df
city_df['Max_temp_F'] = Max_temp_F 

# Find rows that meet the 4 conditions 
city_df2 = city_df[((city_df['Max_temp_F'] > 70) & (city_df['Max_temp_F'] < 80)) 
                   &  (city_df['Wind_Speed'] < 10) & (city_df['Cloudiness'] == 0)]

# Drop any rows will null values. 
city_df2.dropna()

city_df2.head()
```

```python
# Select the necessary fields and save
hotel_df=city_df2[['City', 'Country', 'Lat', 'Lng']]

# Create an empty variable for hotel names
hotel_df['hotel_name']=''
hotel_df
```

```python
# Hotel Map

# Store all of those necessary variables into hotel_df.
# Add a "Hotel Name" column to the DataFrame.
hotel_df=city_df2[['City', 'Country', 'Lat', 'Lng']]

hotel_df['hotel_name']=''

# Set parameters to search for hotels with 5000 meters.
target_type = "Londgin" 
city_coords = city_df2[["Lat", "Lng"]]
radius = 5000

# rewrite params dict

params = {
    "types": target_type,
    "radius": radius,
    "key": g_key
}

for index,row in hotel_df.iterrows():
    lat=row['Lat']
    lng=row['Lng']
    params["location"] = f"{lat},{lng}"

    # Build URL using the Google Maps API
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"


    # Run request
    response = requests.get(base_url, params)

    # print the response URL, avoid doing for public GitHub repos in order to avoid exposing key
    # print(response.url)

    city_hotels = response.json()

    try:
        hotel_df.loc[index,'hotel_name']=city_hotels["results"][0]["name"]
    except (KeyError, IndexError):
        print("missing results")

hotel_df
```

```python
# Plot markers on top of the heatmap.

# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{hotel_name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
marker_layer=gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig
```
