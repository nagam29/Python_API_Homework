```python
# Humidity Heatmap
# Configure gmaps.
# Use the Lat and Lng as locations and Humidity as the weight. 
# Add Heatmap layer to map. 
# heatmap looks different from the example. no satellite view ... color is different etc???? 

#city_df.dtypes

# Store latitude and longitude in locations
locations = city_df[["Lat", "Lng"]]
locations.head()

# Humidity as the weight
weight_humi=city_df['Humidity']

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=weight_humi, 
                               dissipating=False, max_intensity=100,
                                point_radius=4)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
